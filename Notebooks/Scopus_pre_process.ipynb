```python
import pandas as pd
import pycountry


country_list = [country.name for country in pycountry.countries]


#simple first, we use a llm later if needed. this shpuld be enough if the list from python includes those countries
def extract_all_countries(affil):
    affil = str(affil)
    found = []
    for country in country_list:
        if country.lower() in affil.lower():
            found.append(country)
    return list(set(found)) if found else ['Unknown']

data_final['Country'] = data_final['Affiliations'].apply(extract_all_countries)

#to visualize all so we can manually check. i will disbale this later
pd.set_option('display.max_colwidth', None)
data_final.head()
```

I displyed the full text for the collumns so we can check if the country extraction works properrly and manages to match all the countries from the affiliations - and indded it works. There are 531 entries labeled as unknown. We could have the possiblity to drop them, but I will use an **encoder model** to extract those automatically in the cases where the simple string matching. Bert should do the job.

Also one solution would be to actually fine tune a model for this - as we already have the countries extracted correctly for the rest of the entries, we could train a model on this labeled data, but usually this is not needed as it's a simple classification task.

```python
from transformers import pipeline
import torch
import pycountry
from tqdm import tqdm
import pandas as pd

#we will use a smaller version of bert as it is enough for this simple task
device = 0 if torch.cuda.is_available() else -1
print("Using GPU:" if device == 0 else "Using CPU")

classifier = pipeline(
    "zero-shot-classification",
    model="valhalla/distilbart-mnli-12-1",
    device=device
)

#we need to expand a little bit the list of possible abbreviations from the countries list + adding some that are not there

base_countries = [country.name for country in pycountry.countries]
extras = ["USA", "UK", "South Korea", "North Korea", "Russia", "Iran", "Palestine", "Hong Kong", "Taiwan"]
candidate_labels = sorted(list(set(base_countries + extras)))

#the treshold is low, but this is fine/ the actual problem is that most of the time some lables are just abbreivated
def classify_affiliation(affil):
    if not isinstance(affil, str) or affil.strip() == "":
        return ["Unknown"]
    try:
        result = classifier(affil, candidate_labels, multi_label=True)
        labels = [label for label, score in zip(result['labels'], result['scores']) if score > 0.3]
        return labels if labels else ["Unknown"]
    except Exception as e:
        print("Error on:", affil[:80], "-", e)
        return ["Unknown"]

#we do this classification ohnly on the 531 unkown labeled instances. if we manage to still label at least half of them,
#it means this was a sucessful operation :)
mask = data_final['Country'].apply(lambda x: isinstance(x, list) and 'Unknown' in x)

#making a dummy collumn
if 'PredictedCountry' not in data_final.columns:
    data_final['PredictedCountry'] = None


submask = mask & data_final['PredictedCountry'].isna()


affils = data_final.loc[submask, 'Affiliations'].fillna('')
predicted = []

for affil in tqdm(affils, desc="Classifying unknown affiliations"):
    predicted.append(classify_affiliation(affil))

data_final.loc[affils.index, 'PredictedCountry'] = pd.Series(predicted, index=affils.index, dtype='object')

```

After using **distilBERT** (smaller model, easier for our computers) we need to merge the country collumn with the dummy collumn we used for predicting countries for the unknown labeled ones from before.

A bert model specifically trained for **NER** would proabbly be even better. However, the problem we have righ now is just the names normalization. So for further improvements that's the way to go.

```python
def merge_predicted(row):
    if isinstance(row['Country'], list) and 'Unknown' in row['Country']:
        return sorted(list(set([c for c in row['Country'] if c != 'Unknown'] + row['PredictedCountry'])))
    return row['Country']

data_final['Country'] = data_final.apply(merge_predicted, axis=1)


data_final.to_csv("data_final_with_countries.csv", index=False)
```

Now we clean the collumn with the full names of authors. We need to remove the id's from there and the pharanatses. The good part is that all entries follow the same format so we remove anything within pharatheses.

```python
import re

def clean_author_names(author_str):
    if isinstance(author_str, str):
        return re.sub(r'\s*\([^)]*\)', '', author_str)
    return author_str

data_final['Author full names'] = data_final['Author full names'].apply(clean_author_names)

data_final.to_csv("data_processed.csv", index=False)
```

```python
import re

def clean_abstract(text):
    if pd.isna(text):
        return ""

    text = str(text)


    text = re.sub(r'\$.*?\$', '', text)


    text = re.sub(r'\s+', ' ', text)

    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)


    text = text.replace('\xa0', ' ').replace('\u200b', ' ')


    text = re.sub(r'<.*?>', '', text)


    text = text.strip(" .:;-")

    return text.strip()
```

```python
data["Abstract"] = data["Abstract"].apply(clean_abstract)
```
