In this notebook I will create a pre-class for our app to load and pre process the initial 2 datasets:

*   The research papers
*   The patents

So in the end we can select a topic (they both have the same labels) and then see which countries have contirubuted the most to that label, combined.


This means we first have to normalize the countries then we can see that the patents have Europe desribed as a whole. We will seperate it back to specific countries from Europe available in the research section.



```python
import pandas as pd
```

```python
research = pd.read_csv("cleaned_final_data.csv")
patents = pd.read_csv("patents_labeled.csv")
```

```python
import ast
#not sure if this is needed they might have alreadyc been python lists from the previous pre procerssing i did
research["Country"] = research["Country"].apply(ast.literal_eval)

#make them a list as well for patents
patents["Country"] = patents["Country"].apply(lambda x: [x] if isinstance(x, str) else [])
```

Because a label will contain usually mutltiple countries (from the authors affiliations) we will sepearate the same label multime times for different countries so at the end we can sum everything up and group per label and country.

```python
research_exploded = research.explode("Country")
patents_exploded = patents.explode("Country")
```

```python
research_exploded.head()
```

```python
patents_exploded.head()
```

We have country labels called **WIPO** or **PCT** for the lables part in the patents which we will just rename them as international but without modyfing the dataset.


```python
def clean_country(name):
    name = name.strip()
    if "Wipo" in name or "PCT" in name:
        return "International"
    return name

#we apply this on our exploded data without changing the initial datasset
research_exploded["Country"] = research_exploded["Country"].apply(clean_country)
patents_exploded["Country"] = patents_exploded["Country"].apply(clean_country)
```

For each label and cuntry we count how many rows we have. So we just need to group per label and country pairs and then just count. We do this for both datasets.

```python
research_counts = (
    research_exploded.groupby(["Label", "Country"])
    .size()
    .reset_index(name="Research_Count")
)

patent_counts = (
    patents_exploded.groupby(["Label", "Country"])
    .size()
    .reset_index(name="Patent_Count")
)
```

outer join: includes all rows from both DataFrames
then we just sum together

```python
combined = pd.merge(research_counts, patent_counts, on=["Label", "Country"], how="outer")
combined.fillna(0, inplace=True)
```

```python
combined["Total"] = combined["Research_Count"] + combined["Patent_Count"]
```

I notecied some country labeles called ROC for example and after a google search I saw that is the republic of taiwan. I make this function to normalize some more uncommon names. If we encounter more we will add it here.

```python
def normalize_country_name(country):
    if country == "ROC":
        return "Taiwan"
    return country
```

```python
combined["Country"] = combined["Country"].apply(normalize_country_name)
```

```python
#we just want unique countries from the research because it includes all indivdually european countries
valid_countries = set(research_exploded["Country"].unique())
#we update a little bit the countries to have countries just from europe and no "international" for example
filtered_combined = combined[combined["Country"].isin(valid_countries)]

def get_ordered_countries_only(label, top_n=10):

  """
    Returns the top contributing countries for a given research label, ordered by total contributions.
    Only the 'Label' and 'Country' columns are returned  count columns are excluded because we do not care.

    Input: label desired and top countries number

    Output: A DataFrame containing the 'Label' and 'Country' columns for the top N contributors
    """
    filtered = filtered_combined[filtered_combined["Label"] == label]
    sorted_filtered = filtered.sort_values("Total", ascending=False)
    result = sorted_filtered[["Label", "Country"]].head(top_n).reset_index(drop=True)
    return result
```

```python
top_countries = get_ordered_countries_only("quantum computing in ai")
print(top_countries)
```
